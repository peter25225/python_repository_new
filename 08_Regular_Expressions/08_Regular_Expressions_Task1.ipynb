```python
# Задание 1.
# Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером.
# Если да, то функция должна возвращать отдельно номер и регион.
num_plate_corr = 'к727св777' #пример корректного номера
num_plate_incorr = 'к7c7св777' #пример некорректного номера
def checl_num_plate(num_str=''):
# Задаем регулярное выражение:
# - одна буква из заданного списка букв
# - три цифры
# - две буквы из заданного списка букв
# - две или три цифры региона
    check_num_str = r'[авекмнорстух]\d{3}[авекмнорстух]{2}\d{2,3}'
# Отдельно задаем аналогичное регулярное выражение, но с развибкой на номер и регион
    check_num_str_group = r'([авекмнорстух]\d{3}[авекмнорстух]{2})(\d{2,3})'
# Проверяем, что проверяемая строка содержит корректный номер и только его 
    if re.findall(check_num_str,num_str) and re.findall(check_num_str,num_str)[0] == num_str:
# В этом случае возвращаем номер и регион
        return re.findall(check_num_str_group,num_str)
    else:
        print('Номер не валидный')
        return None

print(checl_num_plate(num_plate_corr))
print(checl_num_plate(num_plate_incorr))
    
```

```python
# Задание 2.
# Постарайтесь написать (в крайнем случае найти) самый полный шаблон для проверки e-mail адресов
# через регулярное выражение, которая проверяет e-mail адрес на валидность.
import re

# Используем списки корректных и некорректных e-mail адресов для проверки функции
correct_emails = ['email@example.com',
'firstname.lastname@example.com',
'email@subdomain.example.com',
'1234567890@example.com',
'email@example-one.com',
'email@example.name',
'email@example.museum',
'email@example.co.jp',
'firstname-lastname@example.com']

incorrect_emails = ['plainaddress',
'email@123.123.123.123',
'email@[123.123.123.123]',
'"email"@example.com',
'_______@example.com',
'@%^%#$@#$@#.com',
'@example.com',
'Joe Smith <email@example.com>',
'email.example.com',
'email@example@example.com',
'.email@example.com',
'email.@example.com',
'email..email@example.com',
'あいうえお@example.com',
'email@example.com (Joe Smith)',
'email@example', # не понимаю, почему проверка пропускает его, т.к. у нас явно задано наличие точки после @
'email@-example.com',
'email@111.222.333.44444',
'email@example..com',
'Abc..123@example.com']

def check_email(email_str):
    check_email_str = r'([a-zA-Z0-9]{1,}[a-zA-Z0-9_.-]{2,}[a-zA-Z0-9]{1,})(@)([a-zA-Z0-9]{1,})([a-zA-Z0-9_.-]{1,})(.)([a-zA-Z]{2,4})'
    exclude_str = r'(\.)\1'
   
    if re.fullmatch(check_email_str, email_str) and not re.findall(exclude_str, email_str):
#     if re.fullmatch(check_email_str, email_str):
        print('ok')
    else:
        print('invalid')
# Проверяем список корректных email
for email in correct_emails:
    print(f' {email}', end=' - ')
    check_email(email)

# Проверяем список некорректных email
for email in incorrect_emails:
    print(f' {email}', end=' - ')
    check_email(email)
 
    
    
```

```python
# **Задание 3. **
# Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки
# при помощи регулярных выражений.
import re

dup_str = 'Так значительно быстрее быстрее, чем запускать python python снова и снова, и гораздо проще проще при отладке.'

def del_duplicate_words(check_str):
    return re.sub(r'\b(\w+)( \1\b)+', r'\1', check_str)

del_duplicate_words(dup_str)
```

```python
# Задание 4.
# Напишите функцию, которая будет проверять номер сотового телефона на валидность,
# если он валиден, то переводить его в формат:
# +7(xxx)-xxx-xx-xx
import re

tel_examples = ['1234567890',
'9234567890',
'+7 (923) 456 7890',
'+7 (923) 456 78 90',
'+7 (923) 456 78-90',
'+7 (923) 456-78-90',
'+7 (923) 456-78-90',
'89234567890',
'8(923)4567890',
'8 (923) 456 7890',
'8 (923) 456 78 90',
'8 (923) 456 78-90',
'8 (923) 456-78-90',
'8 (923) 456-78-90',
'8 923 456 7890',
'8 923 456 78 90',
'8 923 456 78-90',
'8 923 456-78-90',
'8 923-456-78-90',
'8-923-456-78-90',
'+79234567890',
'+7(923)4567890',
'+7 923 456 7890',
'+7 923 456 78 90',
'+7 923 456 78-90',
'+7 923 456-78-90',
'+7 923-456-78-90',
'+7-923-456-78-90']


def tel_convert(tel_number):
# Убираем все остальные символы, чтобы проверить количество цифр в строке
    tel_str = re.sub(r'\D','',tel_number)
# Для варианта с 11 цифрами считаем корректным начало '8' или '+7' плюс следующая цифра должна быть '9'
    if len(tel_str) == 11 and (tel_number[0]=='8' or tel_number[0:2] == '+7') and tel_str[1]=='9':
        print(f'+7({tel_str[1:4]})-{tel_str[4:7]}-{tel_str[7:9]}-{tel_str[9:11]}')
# Для варианта с 10 цифрами считаем корректным, если первая цифра '9'
    elif len(tel_str) == 10 and tel_str[0]=='9':       
        print(f'+7({tel_str[0:3]})-{tel_str[3:6]}-{tel_str[6:8]}-{tel_str[8:10]}')
# Иначе корректный номер построить не получится
    else:
        print('Некорректный номер телефона')

for tel_num in tel_examples:
    tel_convert(tel_num)
```
