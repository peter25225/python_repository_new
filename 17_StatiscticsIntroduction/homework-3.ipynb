Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
max_value = max(df['Freedom'])
min_value = min(df['Freedom'])
qty = df['Freedom'].count()
average = df['Freedom'].sum()/qty

print(f'Количество значений: {qty}')
print(f'Минимум: {min_value}')
print(f'Максимум: {max_value}')
print(f'Размах: {max_value-min_value}')
print(f'Медиана: {statistics.median(df.Freedom)}')
print(f'Мат. ожидание: {average}')

df['Freedom'].sort_values() # - заметим, что в одной из строк нет значения Freedom, это оказывает влияние
# на то, что медиана и среднее значение не совпадают, хотя это просто список чисел от 1 до 155
```

```python
# Если удалить строку с отсутствующим значением, мат. ожидаение изменится
df_Freedom = df.drop([28])
max_value = max(df_Freedom['Freedom'])
min_value = min(df_Freedom['Freedom'])
average = df_Freedom['Freedom'].sum()/df_Freedom['Freedom'].count()
print(f'Минимум: {min_value}')
print(f'Максимум: {max_value}')
print(f'Размах: {max_value-min_value}')
print(f'Медиана: {statistics.median(df_Freedom.Freedom)}')
print(f'Мат. ожидание: {average}')
print(f'Стандартное отклонение: {statistics.pstdev(df_Freedom.Freedom)}')
print(f'Дисперсия: {statistics.pvariance(df_Freedom.Freedom)}')
```

Посмотреть корреляцию между Freedom и positive affect (графически и математически). Графически - plt.scatter, математически - посчитать коэффициент Пирсона. О чем говорит знак коэффициента?

```python
freedom = df_Freedom['Freedom']
positive = df_Freedom['Positive affect']
fig, ax = plt.subplots()
ax.scatter(positive, freedom, c='red', alpha=0.3, edgecolors='none')

ax.grid(True)
plt.xlabel('positive')
plt.ylabel('freedom')
plt.title('Точечный график двух случайных величин')
plt.show()

# Видим на диаграмме, что в целом есть прямая корреляция между показателями Freedom и Positive affect

corr, p_value = pearsonr(positive, freedom)
print('Коэффициент корреляции Пирсона:', corr)
# Коэффициент корреляции Пирсона составляет почти 70%, что также свидетельствует о сильной зависимости между ними.
# Знак "+" свидетельствует о прямой корреляции - при увеличении одного показателя растет и другой
```

Нормализовать данные колонки Freedom (централизация + масштабирование)

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

freedom_norm = min_max_scaler.fit_transform(np.array(freedom).reshape(-1, 1))
print(min(freedom_norm))
print(max(freedom_norm))
```

Некоторые выводы:
1) В целом  зависимость между объемами продаж и средней ценой выражена слабо (к-нт Пирсона от -0.17 до -0.20).

2) Однако есть некоторый паттерн зависимости цены от объема - в том числе от объема каждого из вида продукта: 4046, 4225,4770 (заметим, что у них сильная положительная корреляция с Total Volume). При небольших объемах цена может варьироваться очень сильно (это может соответствовать продажам в определенные сезоны, когда предложение не очень большое).
