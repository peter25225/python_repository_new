```python
long_phrase = 'Насколько проще было бы писать программы, если бы не заказчики'
short_phrase = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

print("Length (long):",len(long_phrase))
print("Length (short):",len(short_phrase))

## сравниваем длины фраз, если первая длиннее - выводим True, иначе - False
if len(long_phrase) > len(short_phrase):
    print("Long phrase longer: True")
else:
    print("Long phrase longer: False")
```

```python
## запрашиваем число байт
bytes_num = input("Введите число байт: ")
## проверяем, что введено число
if bytes_num.isdigit():
## если ввод верный - переводим число байт в мегабайты и выводим в нужном формате 
    print("Объем файла равен {:.2f}Mb".format(int(bytes_num)/1048576))
else:
## если ввод неверный - сообщаем об ошибке
    print("Неверный ввод числа")
```

```python
## создаем двухмерный массив для хранения данных о знаках зодиака 
## в 1-й столбец сохраняем название месяца
## во 2-й столбец сохраняем день этого месяца, после которого меняется знак зодиака
## в 3-4 столбцы сохраняем названия знаков для первой и второй части месяца соответственно 
zodiaс = []
zodiaс.append(['январь',19,'козерог','водолей'])
zodiaс.append(['февраль',19,'водолей','рыбы'])
zodiaс.append(['март',21,'рыбы','овен'])
zodiaс.append(['апрель',21,'овен','телец'])
zodiaс.append(['май',21,'телец','близнецы'])
zodiaс.append(['июнь',21,'близнецы','рак'])
zodiaс.append(['июль',21,'рак','лев'])
zodiaс.append(['август',21,'лев','дева'])
zodiaс.append(['сентябрь',21,'дева','весы'])
zodiaс.append(['октябрь',21,'весы','скорпион'])
zodiaс.append(['ноябрь',20,'скорпион','стрелец'])
zodiaс.append(['декабрь',21,'стрелец','козерог'])

## задаем дополнительный одномерный массив для перебора месяцев в цикле
cur_month = []
## задаем новую флажок для отслеживания, был ли найден месяц, с исходным значением '0'
month_found = 0

## запрашиваем месяц и день рождения
month = input("Введите месяц вашего рождения: ")
day = int(input("Введите день вашего рождения: "))

## в цикле от 1 до 12 последовательно выбираем все месяца
for i in range(12):
    cur_month = zodiaс[i]
## при совпадении названия месяца с введенным - проверяем день рождения, а также устанавливаем флаг "месяц найден"
    if cur_month[0] == month:
        month_found = 1
## если день рождения меньше или равен дню смены знака - выводим первый знак месяца, если больше - второй
        if day <= cur_month[1]:
            print("Ваш знак зодиака: " + cur_month[2])
        else:
            print("Ваш знак зодиака: " + cur_month[3])
            
## проверяем, был ли найден месяц, если совпадений не нашлось - сообщаем о неверном вводе названия
if month_found == 0:
    print("Месяц не найден")
```

```python
class StopExecution(Exception):
    def _render_traceback_(self):
        pass

## запрашиваем размер заработной платы
input_str = input("Введите сумму заработной платы в месяц: ")
## проверяем, что введено число
if input_str.isdigit():
    monthly_salary = int(input_str)
## если ввод неверный - сообщаем об ошибке и заканчиваем работу
else:
    print("Неверный ввод числа")
    raise StopExecution
      
## запрашиваем процент затрат на ипотеку
input_str = input("Введите сколько процентов уходит на ипотеку: ")
## проверяем, что введено число
if input_str.isdigit():
    mortgage_share = int(input_str)
## проверяем, что введено число от 0 до 100, если ввод неверный - сообщаем об ошибке и заканчиваем работу
    if mortgage_share > 100 or mortgage_share < 0:
        print("Требуется число в диапазоне от 0 до 100")
        raise StopExecution
## если ввод неверный - сообщаем об ошибке и заканчиваем работу
else:
    print("Неверный ввод числа")
    raise StopExecution

## запрашиваем процент затрат на ипотеку
input_str = input("Введите сколько процентов уходит на жизнь: ")
## проверяем, что введено число
if input_str.isdigit():
    expenses_share = int(input_str)
## проверяем, что введено число от 0 до 100, если ввод неверный - сообщаем об ошибке и заканчиваем работу
    if expenses_share > 100 or expenses_share < 0:
        print("Требуется число в диапазоне от 0 до 100")
        raise StopExecution
## если ввод неверный - сообщаем об ошибке и заканчиваем работу
else:
    print("Неверный ввод числа")
    raise StopExecution

## проверяем, что сумма процентов не выше 100, если ввод неверный - сообщаем об ошибке и заканчиваем работу
if expenses_share + mortgage_share > 100:
    print("Сумма расходов не должна превышать 100%")
    raise StopExecution
    
## запрашиваем число премий в год
input_str = input("Введите количество премий за год: ")
## проверяем, что введено число
if input_str.isdigit():
    bonus_num = int(input_str)
## проверяем, что введено число от 0 до 12
    if bonus_num > 12 or bonus_num < 0:
        print("Требуется число в диапазоне от 0 до 12")
        raise StopExecution
## если ввод неверный - сообщаем об ошибке и заканчиваем работу
else:
    print("Неверный ввод числа")
    raise StopExecution

## производим расчет
mortgage_amount = (monthly_salary*mortgage_share/100)*12
expenses_amount = (monthly_salary*expenses_share/100)*12
savings_amount = monthly_salary*bonus_num/2

print(" ")
print("На ипотеку было потрачено: ",mortgage_amount," рублей")
print("Было накоплено: ", (monthly_salary*12)-mortgage_amount-expenses_amount+savings_amount)
    
```
