```python
# Исходный массив данных - документы
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
    {'type': 'insurance', 'number': '12', 'name': 'Павел Аристархов'},    
]

# Исходный массив данных - полки
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

# Список команд главного меню
commands_main = {
    '1': 'Управление документами',
    '2': 'Управление полками',
    '0': 'Выход'
}

# Список команд для управления документами
commands_documents = {
    '1': 'Найти документ по его номеру',
    '2': 'Создать новый документ',
    '3': 'Удалить документ',
    '4': 'Переместить документ',
    '0': 'Перейти в главное меню'
}

# Список команд для управления полками
commands_directories = {
    '1': 'Создать полку',
    '2': 'Удалить полку',
    '0': 'Перейти в главное меню'
}
```

```python
import time
from IPython.display import clear_output

delay = 0
max_command_num = 4


# Функция поиска документа по номеру
def find_document(document_number=None, print_warning=False, document_print=False):
    if not document_number:
        document_number = input('Введите номер документа для поиска:')
    for document in documents:
        if document.get('number') == document_number:
            if document_print == True:
                print_document(document_number)
            return True
    if print_warning:
        print('Такого документа не существует')
    return False    

# Функция вывода информации о документе
def print_document(document_id):
    for document in documents:
        if document.get('number') == document_id:
            print(document.get('type'),'|',document.get('number'),'|',document.get('name'))

# Функция поиска полки
def find_directory(directory_number):
    for directory in directories:
        if directory == directory_number:
            return True
    return False
            
# Функция создания полки
def create_directory():
    print('')
    directory_number = input('Введите номер полки:')
    if not find_directory(directory_number):
        directories[directory_number] = []
        print('Полка создана')
    else:
        print('Такая полка уже существует')

# Функция удаления полки
def delete_directory():
    print('')
    directory_number = input('Введите номер полки:')
    if find_directory(directory_number):
        del(directories[directory_number])
        print('Полка удалена')
    else:
        print('Полка не найдена')

            
# Поиск номера полки по номеру документа
def find_document_directory(document_number, directory_number=None):
    if find_document(document_number,True):
        if directory_number:
            for directory_document_number in directories.get(directory_number):
                if directory_document_number == document_number:
                    return directory_number
        else:
            for directory in directories:
                for directory_document_number in directories.get(directory):
                    if directory_document_number == document_number:
                        return directory
    return None
            
# Функция поиска документов, не размещенных на полках
def no_directory_documents(print_documents=False):
    if print_documents == False:
        for document in documents:
            if not find_document_directory(document.get('number')):
                return True
        return False        
    else:
        for document in documents:
            if not find_document_directory(document.get('number')):
                print_document(document.get('number'))
            
# Функция вывода списка полок и документов    
def list_directories():
    for directory in directories:
        print('Полка:',directory)
        for document_id in directories.get(directory):
            print('   ',end='')
            print_document(document_id)
    if (no_directory_documents()):
        print('')
        print('Документы без полки:')
        no_directory_documents(True)

# Функция создания документа
def create_document():
    clear_output(wait=False)
    document_dict = {}
    document_number = input('Введите номер документа:')
    if not find_document(document_number):
        document_type = input('Введите тип документа:')
        document_name = input('Введите имя владельца документа:')
        document_dict['type'] = document_type
        document_dict['number'] = document_number    
        document_dict['name'] = document_name
        documents.append(document_dict)
        print('Документ успешно зарегистрирован')
    else:
        print('Такой документ уже зарегистрирован')

# Функция помещения документа на полку
def add_to_directory(document_number=None, directory_number=None, move_document=False):
    if not document_number:
        print('')
        document_number = input('Введите номер документа:')
    if find_document(document_number,True):
        if move_document:
            remove_from_directories(document_number)
        if not directory_number:
            print('')
            directory_number = input('Введите номер полки для размещения документа:')
    if not find_document_directory(document_number,directory_number):
        directories.get(directory_number).append(document_number)
    else:
        print('Документ уже находится на этой полке')

# Функция удаления документа с полок
def remove_from_directories(document_number,print_warning=False):
    if find_document_directory(document_number):
        directories.get(find_document_directory(document_number)).remove(document_number)
    else:
        if print_warning:
            print('Документа нет ни на одной полке')
    
# Функция удаления документа
def delete_document(document_number=None):
    if not document_number:
        print('')
        document_number = input('Введите номер документа для удаления:')
    if find_document(document_number, True):
        for document in documents:
            if document_number == document.get('number'):
                documents.remove(document)
                print('Документ удален')

# Функция запуска меню (в парметр передаем одно из 3-х значений: main - главная, documents - документы, directories - полки)
def run_menu(view='main'):
    clear_output(wait=False)
    list_directories()
    print('- - - - - - - - -')
    print('Список команд:')
    if view == 'documents':
        for command in commands_documents:
            print(command,'-',commands_documents.get(command))
            time.sleep(delay)
        print('- - - - - - - - -')
        get_command(view)
    elif view == 'directories':
        for command in commands_directories:
            print(command,'-',commands_directories.get(command))
            time.sleep(delay)
        print('- - - - - - - - -')
        get_command(view)
    elif view == 'main': 
        for command in commands_main:
            print(command,'-',commands_main.get(command))
            time.sleep(delay)
        print('- - - - - - - - -')
        get_command(view)
    else:
        print('Неверный вызов команды. Завершение работы')

# Функция проверки корректности введенной команды
def check_command(input_str):
    if input_str.isdigit():
        command_num = int(input_str)
        if command_num >= 0 and command_num <= max_command_num:
            return True
        else:
            return False
    else:
        return False

        
# Функция-диспетчер команд 
def get_command(view='main'):
    command_sting = input('Введите команду: ')
    if view == 'documents' and check_command(command_sting):
        run_command_documents(command_sting)
    elif view == 'directories' and check_command(command_sting):
        run_command_directories(command_sting)
    elif view == 'main' and check_command(command_sting):
        run_command_main(command_sting)
    
# Функция для выполнения команд, связанных с управлением документами
def run_command_documents(run_command):
    if run_command == '0':
        run_menu('main')
    elif run_command == '1': # Поиск документа по номеру
        find_document(None,True,True)
        get_command('documents')
    elif run_command == '2': # Создание документа
        create_document()
        time.sleep(0.5)
        run_menu('documents')
    elif run_command == '3': # Удаление документа
        delete_document()
        time.sleep(0.5)
        run_menu('documents')
    elif run_command == '4': # Перемещение документа
        add_to_directory(move_document=True)
        time.sleep(0.5)
        run_menu('documents')        
    else:
        print('')
        print('Неверная команда. Завершение работы')   

# Функция для выполнения команд, связанных с управлением полками
def run_command_directories(run_command='0'):
    if run_command == '0':
        run_menu()
    elif run_command == '1': # Cоздание полки
        create_directory()
        time.sleep(0.2)
        run_menu('directories') 
    elif run_command == '2': # Удаление полки
        delete_directory()
        time.sleep(0.2)
        run_menu('directories') 
    else:
        print('')
        print('Неверная команда. Завершение работы')   

# Функция для выполнения команд главного меню
def run_command_main(run_command='0'):
    if run_command == '0': # Завершение работы
        print('Работа завершена')
    elif run_command == '1': # Меню работы с документами
        run_menu('documents')
    elif run_command == '2': # Меню работы с полками
        run_menu('directories')
    else:
        print('')
        print('Неверная команда. Завершение работы')   
            
run_menu()
```
