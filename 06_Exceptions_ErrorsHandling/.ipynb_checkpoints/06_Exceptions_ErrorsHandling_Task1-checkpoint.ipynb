```python
# Задание 1
# Напишите функцию date_range, которая возвращает список дней между датами start_date и end_date.
# Даты должны вводиться в формате YYYY-MM-DD.
from datetime import datetime
from datetime import timedelta

start_date = '2020-05-14'
end_date = '2020-05-20'

def date_range(start_date,end_date):    
    date_list = []
# Пытаемся сконвертировать строку в дату требуемого формата (дата начала)
    try:
        start_date_dt = datetime.strptime(start_date, '%Y-%m-%d')
    except:
        print('Неверный формат начальной даты')
        start_date_dt = None
# Пытаемся сконвертировать строку в дату требуемого формата (дата конца)
    try:
        end_date_dt = datetime.strptime(end_date, '%Y-%m-%d')
    except:
        print('Неверный формат конечной даты')    
        end_date_dt = None

# Если успешно получили обе даты, итеративно формируем список из дат в заданном диапазоне 
    if start_date_dt and end_date_dt:
        while start_date_dt <= end_date_dt:
            date_list.append(start_date_dt.strftime('%Y-%m-%d'))
            start_date_dt += timedelta(days=1)
    return date_list



date_range(start_date, end_date)
```

```python
# Задание 2
# Дополните функцию из первого задания проверкой на корректность дат.
# В случае неверного формата или если start_date > end_date должен возвращаться пустой список.

start_date = '2020-05-14'
end_date = '2020-05-20'

# Функция проверки периода
def check_dates(start_d,end_d):
# Проверяем, что переменные не пустые
    if not start_d and end_d:
        print('Не задано начало или конец периода')
        return False
# Проверяем тип переданных переменных
    try:
        if type(start_d) == datetime and type(end_d) == datetime:
# Проверяем что дата начала меньше или равна дате конца
            if start_d <= end_d:
                return True
            else:
                print('Начало периода должно быть раньше или равно концу периода')
                return False
        else:
            print('Некорректный тип параметра, необходима дата типа datetime')
            return False
    except:
        return False

# Функция вывода дат заданного периода с проверкой
def date_range_check(start_date,end_date):
    date_list = []

    try:
        start_date_dt = datetime.strptime(start_date, '%Y-%m-%d')
    except:
        print('Неверный формат начальной даты')
        start_date_dt = None
    try:
        end_date_dt = datetime.strptime(end_date, '%Y-%m-%d')
    except:
        print('Неверный формат конечной даты')    
        end_date_dt = None
       
    if check_dates(start_date_dt,end_date_dt):
        while start_date_dt <= end_date_dt:
            date_list.append(start_date_dt.strftime('%Y-%m-%d'))
            start_date_dt += timedelta(days=1)
    return date_list

date_range_check(start_date,end_date)
```
