Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
medv_r = np.array(medv).reshape(-1,1)
lstat_r = np.array(lstat).reshape(-1,1)

model = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(
    medv_r,
    lstat_r,
    test_size = 0.2
)

model.fit(X_train, y_train)

predicted_y = model.predict(X_test)

print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('Стоимость дома')
plt.ylabel('Доля бедного населения')
plt.title('Линейная регрессия от одной переменной')

# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(0, 100, 100)
# Посчитаем предсказанные значения
y = model.coef_*x + model.intercept_
y = np.transpose(y)

plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'g.')
plt.xlim(0,55)
plt.ylim(0,45)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
# data_binom = stats.binom.rvs(n=6,p=0.5,size=10)
# data_binom

import random

def dice_roll(n = 10):
    dice = [1,2,3,4,5,6]
    dice_list = []
    i=0

    while i < n:
        dice_list.append(random.choice(dice))    
        i+= 1;
    return dice_list

sns.distplot(dice_roll(10), bins = 6, kde = False)
plt.title('Плотность вероятности распределения по 10 выпаданий кубика ')
plt.show()

sns.distplot(dice_roll(50), bins = 6, kde = False)
plt.title('Плотность вероятности распределения по 50 выпаданий кубика ')
plt.show()


sns.distplot(dice_roll(1000), bins = 6, kde = False)
plt.title('Плотность вероятности распределения по 1000 выпаданий кубика ')
plt.show()
# Распределение равномерное, так как вероятность выпадения любого числа одинакова
# При увеличении числа экспериментов распределение все больше похоже соответствует теоретическому виду диаграммы
```

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
def dice_roll_sum(cycles_num=1000, rolls=100):
    dice = [1,2,3,4,5,6]
    dice_sum = 0
    i=0
    
    while i<cycles_num:
        dice_sum += random.choice(dice)
        i+=1
    return dice_sum


def dice_roll_sum_list(cycles_num=1000, rolls=100):
    dice_sum_list = []
    i=0
    
    while i<cycles_num:
        dice_sum_list.append(dice_roll_sum(rolls))
        i+=1
    return dice_sum_list

sns.distplot(dice_roll_sum_list(1000,100), bins = 50, kde = False)
plt.title('Плотность вероятности распределения 1000 повторений по 100 подбрасываний кубика ')
plt.show()
# При накоплении сумм результатов подбрасываний кубик видим нормальное распределение
# Центральная предельная теорема применима, так как выполняется ее главное условие:
# в данном эксперименте получаем сумму достаточно большого количества слабо зависимых случайных величин
# (выпадения кубика каждый раз независимы друг от друга)
```
